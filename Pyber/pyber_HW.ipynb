```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

```

```python
# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data= pd.read_csv(city_data_to_load)
ride_data=pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_data=pd.merge(ride_data, city_data, on="city")
combined_data["ride_id"].astype(np.object)
grouped_data=combined_data.set_index("city").groupby(["city"])

date_list=[]

driver_total= grouped_data["driver_count"].mean()


fare=round(grouped_data["fare"].mean(), 2)
ride_total= grouped_data["ride_id"].count()
city_type= city_data.set_index("city")["type"]


uber_data= pd.DataFrame({
    "Drivers Count":driver_total,
    "Type":city_type,
    "Average Fare Per City":fare,
    "Rides Per City":ride_total

})


# Display the data table for preview

uber_data
```

```python
x_axis = np.arange(len(uber_data["Rides Per City"]))
tick_locations = [value for value in x_axis]
```

```python
urban= uber_data.loc[uber_data["Type"]== "Urban"]
rural= uber_data.loc[uber_data["Type"]== "Rural"]
suburban= uber_data.loc[uber_data["Type"]== "Suburban"]

urban_plot = plt.scatter(urban["Rides Per City"], urban["Average Fare Per City"], marker="o", facecolors="lightskyblue", edgecolors="black",linewidth=1,label="urban",
            s=urban["Drivers Count"]*10, alpha=0.75)
rural_plot = plt.scatter(rural["Rides Per City"], rural["Average Fare Per City"], marker="o", facecolors="Gold", edgecolors="black",linewidth=1,label="rural",
            s=rural["Drivers Count"]*10, alpha=0.75)
suburban_plot = plt.scatter(suburban["Rides Per City"], suburban["Average Fare Per City"], marker="o", facecolors="lightcoral", edgecolors= "black", linewidth=1,label="suburban",
            s=suburban["Drivers Count"]*10, alpha=0.75)

plt.xlabel("Total Number Of Rides (Per City)")
plt.ylabel("Average Fare($)")
plt.title("Pyber Ride Sharing Data")
plt.legend(loc="best")
plt.grid()
plt.savefig("Pyber_ride_sharing_data.png")
                       
plt.show()
```

```python
grouped_type_data=combined_data.set_index("type").groupby(["type"])
grouped_type_data.head()
```

```python
total_fares= round(grouped_type_data["fare"].sum(), 2)
print(total_fares)
labels=["rural", "suburban","urban"]
colors = ["Gold", "lightcoral", "lightskyblue"]
explode = (0, 0, 0.1)
plt.pie(total_fares, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Total Fares By City Type")
plt.savefig("Fares_by_city_type.png")
plt.show()
```

```python
total_rides= grouped_type_data["ride_id"].count()
print(total_rides)
labels=["rural", "suburban","urban"]
colors = ["Gold", "lightcoral", "lightskyblue"]
explode = (0, 0, 0.1)
plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Total Rides By City Type")
plt.savefig("rides_by_city_type.png")
plt.show()
```

```python
driver_df= city_data.groupby("type")
total_drivers= driver_df["driver_count"].sum()
print(total_drivers)
labels=["rural", "suburban","urban"]
colors = ["Gold", "lightcoral", "lightskyblue"]
explode = (0, 0, 0.2)
plt.pie(total_drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.title("Total Drivers By City Type")
plt.savefig("drivers_by_city_type.png")
plt.show()
```
